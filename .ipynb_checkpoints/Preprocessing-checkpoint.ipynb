## Eliminate instances where the class value is `NaN`

```python
print('Before NaN elimination: ' + str(df.shape))
df = df.dropna(axis=0, subset
               =[class_ft_renamed])
print('After NaN elimination: ' + str(df.shape))
```

## Assign the right `dtypes` to these features.

```python
for col in df.columns:
    
    # If the column starts with 'discrete' or 'label',
    # make it a 'bool' dtype.
    if col.startswith('b_'):
        # df[col] = df[col].astype(bool)
        df[col] = df[col].astype(float)
    elif col.startswith('s_'):
        df[col] = df[col].astype(str)
    elif col.startswith('i_'):
        df[col] = df[col].astype(int)
    elif col.startswith('f_'):
        df[col] = df[col].astype(float)
    
```

### Train a model

```python
X_dict = ([])
y_dict = ([])

# Dynamically construct DataFrameMapper dictionary based on renamed_ft
for key, value in renamed_ft.items():
    if value.startswith('b_class'):
        # Add class value to special y_dict, apply Binarizer
        entry = ([value],sklearn.preprocessing.Binarizer())
        y_dict.append(entry)
        continue
    elif value.startswith('b_'):
        # Apply Binarizer to all Binary columns
        entry = ([value],sklearn.preprocessing.Binarizer())
        X_dict.append(entry)
    elif value.startswith('f_'):
        # Apply Imputer and StandardScaler to all Numeric columns
        entry = ([value],[sklearn.preprocessing.Imputer(),sklearn.preprocessing.StandardScaler()])
        X_dict.append(entry)        

print X_dict
print y_dict
```

```python
X_mapper = DataFrameMapper(X_dict)
y_mapper = DataFrameMapper(y_dict)

print X_mapper
print y_mapper
```

### Apply the conversion to `np.array`

```python
# print cv.copy().drop(class_ft_renamed,axis=1)

train_set = cv
test_set = dev

X_train = X_mapper.fit_transform(train_set.copy().drop(class_ft_renamed,axis=1))
y_train = y_mapper.fit_transform(train_set.copy()[[class_ft_renamed]])

print X_train[-1].dtype
print y_train[-1].dtype


X_test = X_mapper.fit_transform(test_set.copy().drop(class_ft_renamed,axis=1))
y_test = y_mapper.fit_transform(test_set.copy()[[class_ft_renamed]])
```

```python
print X_train
print y_train
```

```python
# Train the model
model = svm.OneClassSVM()
model.fit(X_train,y_train)
```

```python
# Evaluate Performance
y_pred = model.predict(X_test)
print accuracy_score(y_test, y_pred)
print cohen_kappa_score(y_test, y_pred)
print confusion_matrix(y_test, y_pred)
# print cross_val_score(model, X_train, y_train, cv=4)
```
